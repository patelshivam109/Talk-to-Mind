```python
import pandas as pd
import numpy as np
```

```python
questions = pd.read_csv("../datasets/questionnaire/mental_health_questions.csv")

questions.head()
```

```python
responses = [2] * len(questions)

questions["Response"] = responses

questions.head()
```

Applying reverse scoring

```python
questions["FinalScore"] = questions.apply(
    lambda row: 4 - row["Response"]
    if row["ReverseScore"] == "Yes"
    else row["Response"],
    axis=1
)

questions
```

```python
category_scores = (
    questions
    .groupby("Category")["FinalScore"]
    .sum()
    .reset_index()
)

category_scores
```

```python
overall_score = questions["FinalScore"].sum()

print("Overall Score:", overall_score)
```

Risk levels

```python
if overall_score <= 20:
    risk = "Low"

elif overall_score <= 40:
    risk = "Moderate"

elif overall_score <= 60:
    risk = "High"

else:
    risk = "Very High"

print("Risk Level:", risk)
```

```python
import os

os.makedirs("../reports", exist_ok=True)

category_scores.to_csv(
    "../reports/questionnaire_scores.csv",
    index=False
)

summary = pd.DataFrame({
    "OverallScore": [overall_score],
    "RiskLevel": [risk]
})

summary.to_csv(
    "../reports/questionnaire_summary.csv",
    index=False
)
```
